```python
# Install Excel library
!pip install openpyxl

from openpyxl import Workbook
from openpyxl.chart import LineChart, Reference

# Create workbook
wb = Workbook()

# -------------------
# DASHBOARD SHEET
# -------------------
ws1 = wb.active
ws1.title = "Dashboard"

ws1["A1"] = "Ultimate Integrated Dam Model"
ws1["A3"] = "User:"
ws1["B3"] = "Enter Project Name"

# -------------------
# INPUT SHEET
# -------------------
ws2 = wb.create_sheet("Inputs")

ws2["A1"] = "Parameter"
ws2["B1"] = "Value"

ws2["A2"] = "Catchment Area (km2)"
ws2["B2"] = 125

ws2["A3"] = "Longest Stream Length (km)"
ws2["B3"] = 24.5

ws2["A4"] = "Slope (m/km)"
ws2["B4"] = 4.2

ws2["A5"] = "PMP (cm)"
ws2["B5"] = 65

# -------------------
# SUH SHEET
# -------------------
ws3 = wb.create_sheet("SUH")

ws3.append(["Time(hr)", "Discharge"])

suh = [0,12,38,68,75,54,36,23,14,8,4,2,1,0]

for i, q in enumerate(suh):
    ws3.append([i, q])

# -------------------
# PMF SHEET
# -------------------
ws4 = wb.create_sheet("PMF")

ws4.append(["Time", "Rain", "Runoff"])

for t in range(20):
    rain = 10 if t < 5 else 0

    q = suh[t] if t < len(suh) else 0

    runoff = rain * q

    ws4.append([t, rain, runoff])

# -------------------
# CHART
# -------------------
chart = LineChart()
chart.title = "PMF Hydrograph"

data = Reference(ws4,
                 min_col=3,
                 min_row=1,
                 max_row=21)

cats = Reference(ws4,
                 min_col=1,
                 min_row=2,
                 max_row=21)

chart.add_data(data, titles_from_data=True)
chart.set_categories(cats)

ws4.add_chart(chart, "E2")

# -------------------
# SAVE
# -------------------
filename = "Ultimate_Dam_Model.xlsx"

wb.save(filename)

print("Workbook created:", filename)
```
